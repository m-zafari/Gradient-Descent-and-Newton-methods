```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Gradient Descent

```python
# finds a minimum of f(x)
def gradient_descent(f, df, x0, alpha=5e-2, epsilon=1e-5, max_iter=300):
    xns = []
    xn = x0
    for n in range(0, max_iter):
        xns.append(xn)
        dfxn = df(xn)

        if (abs(dfxn) < epsilon):
            print("Found a minimum after ", n, " iterations")
            #print(xns[-1])
            return xns

        xn = xn - alpha * dfxn

    print('Exceeded maximum iterations. No minimum found.')
    return xns
```

```python
F = lambda x: 2*x*np.cos(x) - x**2 * np.sin(x)
dF = lambda x: 2 * np.cos(x) - 4 * x * np.sin(x) - x**2 * np.cos(x)
```

```python
x = np.linspace(-3, 1.5, 100)

fig = plt.figure()
fig.set_dpi(150)
fig.set_size_inches(6, 4)
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')

plt.plot(x, F(x), 'r');
plt.plot(x, dF(x), 'g');
plt.legend(['f(x)', 'df(x)']);
```

```python
xns = gradient_descent(F, dF, -2)
```

```python
x = np.linspace(-3, 2.5, 100)

fig = plt.figure()
fig.set_dpi(150)
fig.set_size_inches(6, 4)
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')

plt.plot(x, F(x), 'r');
plt.scatter(xns, F(np.array(xns)), marker='X');
plt.legend(['f(x)', 'f(Xn)']);
```

## Newton

```python
#  finds a solution to f(x)=0
def newton(f, df, x0, epsilon=1e-9, max_iter=200):
    xns = []
    xn = x0
    for n in range(0, max_iter):
        xns.append(xn)
        fxn = f(xn)

        if abs(fxn) < epsilon:
            print('Found solution after ', n, ' iterations.')
            return xns

        dfxn = df(xn)

        if dfxn == 0:
            print('Zero derivative. No solution found.')
            return xns

        xn = xn - (fxn / dfxn)

    print('Exceeded maximum iterations. No solution found.')
    return xns
```

```python
G = lambda x: x**3 - x**2 - 1
dG = lambda x: 3*x**2 - 2*x
```

```python
x = np.linspace(-1.5, 2.5, 100)

fig = plt.figure()
fig.set_dpi(150)
fig.set_size_inches(6, 4)
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')

plt.plot(x, G(x), 'r');
plt.plot(x, dG(x), 'g');
plt.legend(['g(x)', 'dg(x)']);
```

```python
xns = newton(G, dG, 2)
```
